```python
# This function should prepare the data by reading it from a file and converting it into a useful format for training and testing

#Constant variable
FIRST_COL_INDEX = 0

def preprocess(filename, info, nbins, discrete_type):
    #get the file information
    
    df = pd.read_csv(filename, header = None, names = info.attributes_names)
    df.replace('?', np.NaN, inplace=True)
    #remove first col index from file
    if(info.isFirstColId == True):
        df.drop(df.columns[FIRST_COL_INDEX], axis=1, inplace = True)
    
    #loops through every column to check if data needs discretizing or not
    for i in range(len(df.columns)):
        if info.attributes_type[i] == 2:
            df.iloc[:,i].astype(float, copy = False, errors = "ignore")
            if(nbins != 0):
                if(discrete_type == "equalwidth"):
                    #equal width discretizing
                    df.iloc[:,i] = pd.cut(df.iloc[:,i], nbins)
                    
    return df
```

```python
# This function should calculat prior probabilities and likelihoods from the training data and using
# them to build a naive Bayes model

def train(train_data, k, info, nbins):
    
    class_data = train_data.iloc[:,info.classindex]
    
    #prior variable as dictionary
    priors = {}
    
    #likelihood dictionary, likelihood[class label][attribute name][attribute value] = P(All possible attribute values | class label)
    likelihood = defaultdict(lambda: defaultdict(lambda: defaultdict(float)))
    
    #converts attributes_type array
    attributes_type = correct_attributes_type(info.attributes_type, nbins)
    
    for label in class_data.unique():
        priors[label] = len(train_data.loc[class_data == label]) / len(train_data)
        
        index = 0
        for attribute_name in train_data.columns:
            if(index == info.classindex):
                #skips if index is the class index
                index += 1
                continue
            
            #number of unique values in the attribute
            d = len(train_data.loc[:,attribute_name].dropna().unique())
            class_name = train_data.columns[info.classindex]
            
            #get the data with the current class
            curr_data = train_data.loc[class_data == label, [attribute_name, class_name]]
            
            
            #if the attribute is non numeric
            if(attributes_type[index] != 2):
                
                #dictionary to store the count of every possible value in the attribute with that class label
                attr_val_count = defaultdict(int)
                
                #Find the count of all possible values in attribute name
                for attr_value in curr_data[attribute_name]:
                     if attr_value != np.NaN:
                        attr_val_count[attr_value] += 1
                
                for data in attr_val_count.keys():
                    #Likelihood for every possible attribute value, P(attribute vals | class label)
                    likelihood[label][attribute_name][data] = (attr_val_count[data] + k)/ (len(curr_data.dropna()) + k*d)
            
            #if attribute is numeric, use gaussina
            else:
                mu = float(curr_data[attribute_name].mean())
                sd = float(curr_data[attribute_name].std())
                likelihood[label][attribute_name] = {"mu": mu, "sd" : sd}   
            index += 1
            
        model = {"priors": priors, "likelihood" : likelihood}
    return model

```

```python
# This function should predict classes for new items in a test dataset (for the purposes of this assignment, you
# can re-use the training data as a test set)

def predict(test_set, model, info, nbins):
    priors = model["priors"]
    likelihood = model["likelihood"]
    
    n = len(test_set)
    
    #drop class for testing
    test_data = test_set.drop(test_set.columns[info.classindex], axis=1)
    
    attribute_names = test_data.columns
    
    attributes_type = correct_attributes_type(info.attributes_type, nbins)
    
    attributes_type.pop(info.classindex)
                         
    prediction = {}
    label_predicted = {}
    
    #i'th value of the data for iloc
    ith = 0
    
    #for every index in the test set
    for index in (test_data.index):
        row = test_data.iloc[ith,:]
        for label in priors.keys():
            cumm_prob = priors[label]
            i = 0
            for attribute in attribute_names:
                if(row[attribute] != np.NaN):
                    #if not numeric , use likelihood
                    if(attributes_type[i] != 2):
                        cumm_prob *= likelihood[label][attribute][row[attribute]]
                    #if numeric, calculates likelihood from given mu and sd
                    else:
                        x = row[attribute] 
                        mu = likelihood[label][attribute]["mu"]
                        sd = likelihood[label][attribute]["sd"]
                        if(mu > 0 and sd > 0):
                            pdf = normal_pdf(x,mu,sd)
                            cumm_prob *= pdf
                i+=1
            prediction[label] = cumm_prob
        
        #get maximum likelihood
        label_predicted[index] = max(prediction, key = prediction.get)
        
        ith += 1
    return label_predicted

```

```python
# This function should evaliate the prediction performance by comparing your model’s class outputs to ground
# truth labels

def evaluate(test_set, classindex, label_predicted):
    true_labels = test_set.iloc[:,classindex]
    result = []
    for index in test_set.index:
        if(true_labels[index] == label_predicted[index]):
            result.append(1)
        else:
            result.append(0)
 
    return result
```

```python
#make testing function, this function uses random subsampling of N size samples. It is possible to alter discretize level and add-k leve

def test_script(filenames, nbins, k, N, discrete_type):
    print("Information:")
    print("########################################")

    if(nbins == 0):
        print("Discretization technique: None")
        print("Level: {}".format(nbins))
    else:
        print("Discretization technique: Equal Width")
        print("Level: {}".format(nbins))
    print("Probabilistic smoothing technique: Add-k")
    print("k = {}".format(k))
    print("Train Test technique: Randomly seperate the data frame, 75% to train , 25% test ")
    #loop through all data
    result_dict = defaultdict(float)
    time_taken = defaultdict(float)
    print("Repition: {}".format(N))
    print("########################################")
    print("Data Name and Average Accuracy:")
    print("########################################")
    
    for i in range(N):
        for filename in filenames:

            info = get_fileInfo(filename)

            df = preprocess(filename, info, nbins, discrete_type)
            msk = np.random.rand(len(df)) < 0.75

            train_set = df[msk]
            test_set = df[~msk]
            
            time_start = time.time()
            model = train(train_set, k, info, nbins)
            time_taken[filename] += time.time() - time_start

            prediction = predict(test_set, model, info, nbins)
            result = evaluate(test_set, info.classindex, prediction)
    
            accuracy = 100 * sum(result)/len(result)
            result_dict[filename] += accuracy
    
    result = {}
    time_result = {}
    for i in result_dict.keys():
        print(i + " accuracy score: " + str(round(result_dict[i]/N,2)) + " %")
        print("TIME TAKEN: " + str(round(time_taken[i]/N, 2)) + " seconds")
        result[i] = round(result_dict[i]/N,2)
        time_result[i] = round(time_taken[i]/N,2)
    print("########################################")
    
    return {"accuracy": result, "time": time_result}
```
